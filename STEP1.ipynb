```python
import pandas as pd
```

```python
reviews = pd.read_csv('/content/amazon_reviews (1).txt',sep='\t')
reviews
```

```python
ndf=reviews.groupby('PRODUCT_CATEGORY')
ndf.groups
```

```python
books_df = ndf.get_group('Books')
Electronics_df = ndf.get_group('Electronics')
Grocery_df = ndf.get_group('Grocery')
```

```python
from google.colab import files
books_df.to_csv('Rbooks_df.csv')
files.download('Rbooks_df.csv')
Electronics_df.to_csv('RElectronics_df.csv')
files.download('RElectronics_df.csv')
Grocery_df.to_csv('RGrocery_df.csv')
files.download('RGrocery_df.csv')
```
